```python
# Load data
import pandas as pd
df = pd.read_csv("../data/raw/data.csv", encoding="ISO-8859-1")
print(df.shape)
```

```python
# Create a copy
clean_df = df.copy()
```

```python
# Remove duplicates
clean_df = clean_df.drop_duplicates()
print(clean_df.shape)
```

```python
# Handle missing Customer IDs
clean_df = clean_df.dropna(subset=["CustomerID"])
clean_df.isnull().sum()
```

```python
# Remove returns
clean_df = clean_df[clean_df["Quantity"] > 0]
(clean_df["Quantity"] < 0).sum()
```

```python
# Remove invalid prices
clean_df = clean_df[clean_df["UnitPrice"] > 0]
```

```python
# Convert date column
clean_df["InvoiceDate"] = pd.to_datetime(clean_df["InvoiceDate"])
clean_df.info()
```

```python
# Create Revenue column
clean_df["Revenue"] = (
    clean_df["Quantity"] *
    clean_df["UnitPrice"]
)
clean_df[["Quantity", "UnitPrice", "Revenue"]].head()
```

```python
# Create time features
clean_df["Year"] = clean_df["InvoiceDate"].dt.year
clean_df["Month"] = clean_df["InvoiceDate"].dt.month
clean_df["Month"] = clean_df["InvoiceDate"].dt.month_name()
```

```python
# Final validation
clean_df.info()
clean_df.isnull().sum()
clean_df.describe()
```

```python
# Save cleaned data
clean_df.to_csv(
    "../data/processed/cleaned_data.csv",
    index=False
)
```
